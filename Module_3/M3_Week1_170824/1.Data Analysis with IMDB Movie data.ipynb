# *8. View missing values*

# *9. Deal with missing values- Deleting*

```python
 # Use drop function to drop columns
data.drop('Metascore', axis=1).head()
# Lệnh trên vẫn chưa drop data thực trên server cho tới khi ta thêm inplace=True.
# axis = 1 là hàng 
```

```python
# Delete rows
data.drop(1, axis=0).head()
```

```python
# Xóa các hàng có Title là "Unknown"
data_cleaned = data[data['Title'] != 'Unknown']
print(data_cleaned.head())
```

```python
# Xóa các hàng có Rating nhỏ hơn 5
data_cleaned = data[data['Rating'] >= 5]
print(data_cleaned.head())
```

```python
# Xóa các hàng có Rating nhỏ hơn 5 và Revenue (Millions) nhỏ hơn 50
data_cleaned = data[(data['Rating'] >= 5) | (data['Revenue (Millions)'] >= 50)]
print(data_cleaned.head())
```

```python
# Xóa các cột "Description" và "Actors"
data_cleaned = data.drop(['Description', 'Actors'], axis=1)
print(data_cleaned.head())
```

```python
# Lọc các cột không có kiểu dữ liệu là object
data_cleaned = data.select_dtypes(exclude=['object'])
print(data_cleaned.head())
```

```python
print(data.columns)
print(type(data.columns)) # cho truy cập phần tử nhưng không gán được như list, có thể đổi sang list
print(data.columns.tolist())
```

```python
# Lọc bỏ các cột có tên chứa từ "Revenue"
data_cleaned = data.drop([col for col in data.columns if 'Revenue' in col], axis=1)
print(data_cleaned.head())
```

```python
# mặc định axis 0 là hàng 
# Xóa các hàng chứa giá trị NaN
df_dropped = data.dropna()  #axis = 0
print(df_dropped.head())
```

```python
# axis 0 row
# axis 1 column 
df_dropped = data.dropna(axis=1)  
df_dropped.head
df_dropped.columns
```
