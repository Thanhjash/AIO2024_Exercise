```python
import numpy as np

def create_train_data_iris():
    data = np.loadtxt("iris.data.txt", delimiter=",", dtype=str)
    return data

print(create_train_data_iris())
```

```python
def train_gaussian_naive_bayes(train_data):
    features = train_data[:, :4].astype(float)
    labels = train_data[:, 4]
    
    classes = np.unique(labels)
    n_classes = len(classes)
    n_features = features.shape[1]
    
    prior_probability = np.zeros(n_classes)
    conditional_probability = np.zeros((n_classes, n_features, 2))
    
    for i, c in enumerate(classes):
        class_data = features[labels == c]
        prior_probability[i] = len(class_data) / len(features)
        conditional_probability[i, :, 0] = np.mean(class_data, axis=0)
        conditional_probability[i, :, 1] = np.var(class_data, axis=0)
    
    return prior_probability, conditional_probability
```

```python
def prediction_iris(X, prior_probability, conditional_probability):
    n_classes = len(prior_probability)
    log_likelihood = np.zeros(n_classes)
    
    for i in range(n_classes):
        log_prior = np.log(prior_probability[i])
        log_conditional = -0.5 * np.sum(np.log(2 * np.pi * conditional_probability[i, :, 1]))
        log_conditional -= 0.5 * np.sum((X - conditional_probability[i, :, 0])**2 / conditional_probability[i, :, 1])
        log_likelihood[i] = log_prior + log_conditional
    
    return np.argmax(log_likelihood)
```

```python
# Ví dụ 1
X = [6.3, 3.3, 6.0, 2.5]
train_data = create_train_data_iris()
y_unique = np.unique(train_data[:, 4])
prior_probability, conditional_probability = train_gaussian_naive_bayes(train_data)
pred = y_unique[prediction_iris(X, prior_probability, conditional_probability)]
print(f"Dự đoán cho {X}: {pred}")
assert pred == "Iris-virginica"

# Ví dụ 2
X = [5.0, 2.0, 3.5, 1.0]
pred = y_unique[prediction_iris(X, prior_probability, conditional_probability)]
print(f"Dự đoán cho {X}: {pred}")
assert pred == "Iris-versicolor"

# Ví dụ 3
X = [4.9, 3.1, 1.5, 0.1]
pred = y_unique[prediction_iris(X, prior_probability, conditional_probability)]
print(f"Dự đoán cho {X}: {pred}")
assert pred == "Iris-setosa"
```
