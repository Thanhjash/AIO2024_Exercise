```python
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
print(os.getcwd())
```

```python
#12

# Đọc ảnh
img = mpimg.imread('dog.jpeg')

```

```python
# Áp dụng phương pháp Lightness
gray_img_01 = (np.max(img, axis=2) + np.min(img, axis=2)) / 2

# Kiểm tra giá trị tại vị trí [0, 0]
print(gray_img_01[0, 0])

# Hiển thị ảnh grayscale Lightness
plt.imshow(gray_img_01, cmap='gray')
plt.title('Grayscale - Lightness')
plt.axis('off')  # Tắt hiển thị trục
plt.show()
```

```python
# Áp dụng phương pháp Average
gray_img_average = np.mean(img, axis=2)

# Kiểm tra giá trị tại vị trí [0, 0]
print(gray_img_average[0, 0])
# Hiển thị ảnh grayscale Average
plt.imshow(gray_img_average, cmap='gray')
plt.title('Grayscale - Average')
plt.axis('off')  # Tắt hiển thị trục
plt.show()
```

```python
# Áp dụng phương pháp Luminosity
gray_img_luminosity = 0.21 * img[:,:,0] + 0.72 * img[:,:,1] + 0.07 * img[:,:,2]


# Kiểm tra giá trị tại vị trí [0, 0]
print(gray_img_luminosity[0, 0])

# Hiển thị ảnh grayscale Luminosity
plt.imshow(gray_img_luminosity, cmap='gray')
plt.title('Grayscale - Luminosity')
plt.axis('off')  # Tắt hiển thị trục
plt.show()
```
